```python
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
import imageio

USAFgreen_path = r"C:\Users\sann7609\Documents\Oxford\Thesis\images\CAM1-green_USAF.png"
USAF_IR_path = r"C:\Users\sann7609\Documents\Oxford\Thesis\images\CAM2-red.png"
```

```python
pixsize = 1.06e-6

USAF_IR    = np.flip(np.flip(imageio.v2.imread(USAF_IR_path),  0),1)[320:690,650:1250] 
USAF_green = np.flip(imageio.v2.imread(USAFgreen_path),0)[350:720,300:900]   # shape: (rows, cols)

fig, ax = plt.subplots(1,2,figsize=(8,6),gridspec_kw={'wspace': 0.05})
ax[0].imshow(USAF_IR, cmap='gray', origin='lower',extent=[0,USAF_IR.shape[1]*pixsize*1e6,0,USAF_IR.shape[0]*pixsize*1e6])
ax[1].imshow(USAF_green, cmap='gray', origin='lower',extent=[0,USAF_green.shape[1]*pixsize*1e6,0,USAF_green.shape[0]*pixsize*1e6])

ax[0].text(s='1030 nm', x=140, y=350, color='red', size=14)
ax[1].text(s='515 nm', x=140, y=350, color='lime', size=14)

ax[1].set_yticklabels([])
ax[0].set_xlabel('x (µm)', size=12)
ax[0].set_ylabel('y (µm)', size=12)
ax[1].set_xlabel('x (µm)', size=12)

plt.savefig(r'C:\Users\sann7609\Documents\Oxford\Thesis\images\USAF_resolution.png', dpi=300, bbox_inches='tight')
```
