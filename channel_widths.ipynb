```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as spl
from scipy.interpolate import interp1d
import abel

# ----- Parameters -----
r_min, r_max = -100e-6, 100e-6
n_r = 1000                 # original desired sampling
wavelength = 1030e-9

# Make image size odd to avoid quadrant/centering issues in pyabel
if n_r % 2 == 0:
    N = n_r - 1   # make odd (999)
else:
    N = n_r

# Recompute radial sampling so physical range matches approximately
# We'll set the pixel spacing so that coords span roughly r_min..r_max
dr = (r_max - r_min) / (n_r - 1)   # keep the original spacing for radial values
coords = (np.arange(N) - N//2) * dr

# Build the 1D radial coordinate (same physical spacing used earlier)
rvals = np.linspace(r_min, r_max, n_r)
# 1D Bessel^2 on full range (signed rvals)
I2_1d = spl.jv(0, 2 * np.pi * alpha_axi * rvals / wavelength) ** 2

# Positive radius axis (including r = 0)
rpos = np.abs(rvals[n_r//2:])   # length ~ n_r/2
I2_pos = spl.jv(0, 2 * np.pi * alpha_axi * rpos / wavelength) ** 2

# Interpolator from radius -> intensity (will return 0 outside range)
interp_I2 = interp1d(rpos, I2_pos, bounds_error=False, fill_value=0.0)

# Create 2D radius grid (square image of size N x N)
X, Y = np.meshgrid(coords, coords)
Rgrid = np.sqrt(X**2 + Y**2)

# Evaluate axisymmetric image
img = interp_I2(Rgrid)

# Ensure no NaNs / infinities
img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0).astype(float)

# ----- PyAbel forward transform (projection) -----
# Let pyabel choose defaults for centering; do not pass symmetry_axis that might be version-sensitive
tr = abel.Transform(img, method='hansenlaw', direction='forward')
proj_img = tr.transform

# Extract the central horizontal row as 1D projection (line-of-sight)
center = proj_img.shape[0] // 2
proj_1d = proj_img[center, :]

# Build x-axis in meters that matches coords
x_coords = coords

# Normalize for plotting
I2_pos_full = np.abs(spl.jv(0, 2 * np.pi * alpha_axi * np.abs(rvals) / wavelength) ** 2)
I2_norm = I2_pos_full / np.max(I2_pos_full)
proj_norm = proj_1d / np.max(proj_1d)

# ----- Plot -----
plt.figure(figsize=(8,5))
# original radial profile (plotted vs signed r from -max..+max)
plt.plot(np.abs(rvals) * 1e6, I2_norm, color='red', label=r'Bessel$^2$ (radial, |r|)')

# projection vs x-position
plt.plot(x_coords * 1e6, proj_norm, color='blue', label='Abel forward (projection)')

plt.xlim(-100, 100)
plt.xlabel('position (μm)')
plt.ylabel('Normalized intensity (arb. units)')
plt.legend()
plt.title('Bessel^2 radial profile and its Abel projection (pyabel)')
plt.grid(True)
```

```python
def get_I_axi_peak( E_axi: float = 1000e-3,  # Energy in the axicon pulse [J]
                    D_beam: float = 60e-3,  # Beam diameter incident on the axicon [m]
                    d_axi: float = 0,       # Diameter of the inner hole in the annulus [m]
                    theta_base_deg: float = 8,  # base angle of the axicon [deg]
                    lambda_beam: float = 1030e-9,  # Beam central wavelength [m]
                    tau_FWHM: float = 900e-15,  # Pulse FWHM duration [s]
                    z: float = 0,  # Start position [m]
                    ):
    # Assumes all energy is in an annulus (outer diameter D_beam, inner hole d_axi)and that the pulse is Gaussian in time but flat-top transverse:

    theta_base_rad = theta_base_deg*np.pi/180
    eta_axi = 1.45 #refractive index of the axicon at 1030nm
    alpha_axi = np.arcsin(eta_axi*np.sin(theta_base_rad))-theta_base_rad #approach angle of the axicon
    I0_a = (1/(1 - (d_axi/D_beam)*2)) * 2*np.sqrt(np.log(2))/(np.sqrt(np.pi)) * E_axi / ((np.pi*D_beam/2)*2 * tau_FWHM)  # Peak intensity [W/m^2] on axicon

    I_onax_z = 4*np.pi*2/lambda_beam * z * I0_a * alpha_axi*2 * 1e-4  # On-axis intensity at location Z [W/cm^2]
    return I_onax_z  # Return distance in mm and HALF the peak intensity [W/cm²]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- your original arrays & params ---
E_BigMama_mJ = [5.1,7.4,11.4,18.8,34,44.7,59.3,81.0,111.5,156,229,339,514,800,885,999,1100,1230,1365,1510,1670,1840,2025,2215,2410,2600,2790,3010]
BM_delays = np.array([1000,950,900,850,800,780,760,740,720,700,680,660,640,620,615,610,605,600,595,590,585,580,575,570,565,560,555,550])
throughput = 0.62
z_probe = 336e-3  # meters

# --- linear interpolation with linear extrapolation helper ---
def linear_interp_extrap(x, xp, yp):
    sort_idx = np.argsort(xp)
    xp_s = np.array(xp)[sort_idx]
    yp_s = np.array(yp)[sort_idx]
    x_arr = np.atleast_1d(x).astype(float)
    y = np.interp(x_arr, xp_s, yp_s)
    # low extrapolate
    if np.any(x_arr < xp_s[0]):
        slope_low = (yp_s[1] - yp_s[0]) / (xp_s[1] - xp_s[0])
        mask_low = x_arr < xp_s[0]
        y[mask_low] = yp_s[0] + slope_low * (x_arr[mask_low] - xp_s[0])
    # high extrapolate
    if np.any(x_arr > xp_s[-1]):
        slope_high = (yp_s[-1] - yp_s[-2]) / (xp_s[-1] - xp_s[-2])
        mask_high = x_arr > xp_s[-1]
        y[mask_high] = yp_s[-1] + slope_high * (x_arr[mask_high] - xp_s[-1])
    return y if y.shape != () else float(y)

# --- YOUR get_I_axi_peak (exactly as you supplied) ---
def get_I_axi_peak( E_axi: float = 1000e-3,  # Energy in the axicon pulse [J]
                    D_beam: float = 60e-3,  # Beam diameter incident on the axicon [m]
                    d_axi: float = 0,       # Diameter of the inner hole in the annulus [m]
                    theta_base_deg: float = 8,  # base angle of the axicon [deg]
                    lambda_beam: float = 1030e-9,  # Beam central wavelength [m]
                    tau_FWHM: float = 900e-15,  # Pulse FWHM duration [s]
                    z: float = 0,  # Start position [m]
                    ):
    # Assumes all energy is in an annulus (outer diameter D_beam, inner hole d_axi)and that the pulse is Gaussian in time but flat-top transverse:

    theta_base_rad = theta_base_deg*np.pi/180
    eta_axi = 1.45 #refractive index of the axicon at 1030nm
    alpha_axi = np.arcsin(eta_axi*np.sin(theta_base_rad))-theta_base_rad #approach angle of the axicon
    I0_a = (1/(1 - (d_axi/D_beam)*2)) * 2*np.sqrt(np.log(2))/(np.sqrt(np.pi)) * E_axi / ((np.pi*D_beam/2)*2 * tau_FWHM)  # Peak intensity [W/m^2] on axicon

    I_onax_z = 4*np.pi*2/lambda_beam * z * I0_a * alpha_axi*2 * 1e-4  # On-axis intensity at location Z [W/cm^2]
    return I_onax_z  # Return distance in mm and HALF the peak intensity [W/cm²]

# --- wrapper: map BM_delay -> E_on_axicon and compute intensity via your function ---
def energy_and_focus_intensity(BM_delay,
                               BM_delays_array=BM_delays,
                               E_BigMama_mJ_array=E_BigMama_mJ,
                               throughput=throughput,
                               z_probe=z_probe,
                               **get_I_kwargs):
    """
    Input:
      BM_delay : scalar or array-like of BM delay values (same units as BM_delays_array)
    Returns:
      E_axi_J   : energy on axicon in J (scalar or np.array matching input)
      I_focus   : intensity returned by get_I_axi_peak (W/cm^2) (scalar or np.array)
    Any additional kwargs are forwarded to get_I_axi_peak (e.g., D_beam, d_axi, theta_base_deg, ...)
    """
    BM_delay_arr = np.atleast_1d(BM_delay).astype(float)
    # interpolate energy in mJ
    E_mJ_interp = linear_interp_extrap(BM_delay_arr, BM_delays_array, E_BigMama_mJ_array)
    # convert to J and apply throughput
    E_axi_J = E_mJ_interp * 1e-3 * throughput

    # compute intensity(s) using your function exactly
    I_list = []
    for E in np.atleast_1d(E_axi_J):
        I_val = get_I_axi_peak(E_axi=E, z=z_probe, **get_I_kwargs)
        I_list.append(float(I_val))
    I_arr = np.array(I_list)

    # return scalar if scalar input
    if np.isscalar(BM_delay):
        return float(E_axi_J), float(I_arr[0])
    else:
        return E_axi_J, I_arr

# --- Example usage ---
if __name__ == "__main__":
    # single delay
    bm = 582.0
    E_axi, I_wc2 = energy_and_focus_intensity(bm)
    print(f"BM delay {bm} -> E_on_axicon = {E_axi:.4e} J, I_focus (per your fn) = {I_wc2:.4e} W/cm^2")

    # multiple delays -> quick plot
    delays = np.linspace(550, 1000, 150)
    E_vals, I_vals = energy_and_focus_intensity(delays)
    plt.figure()
    plt.plot(delays, E_vals, label="E_on_axicon (J)")
    plt.xlabel("BM delay")
    plt.ylabel("Energy (J)")
    plt.legend(loc='upper left')
    ax2 = plt.twinx()
    ax2.plot(delays, I_vals, '--', label="I_focus (W/cm^2)")
    ax2.set_ylabel("Intensity (W/cm^2)")
    ax2.legend(loc='upper right')
    plt.show()
```
